```python
class FilaCircular:
    def __init__(self, capacidade):
        self.capacidade = capacidade
        self.dados = [None] * capacidade
        self.inicio = 0
        self.fim = 0
        self.total = 0

    def esta_vazia(self):
        return self.total == 0

    def esta_cheia(self):
        return self.total == self.capacidade

    def inserir(self, valor):
        if self.esta_cheia():
            print("Fila cheia. Não é possível inserir.")
            return False
        self.dados[self.fim] = valor
        self.fim = (self.fim + 1) % self.capacidade
        self.total += 1
        return True

    def remover(self):
        if self.esta_vazia():
            print("Fila vazia. Não é possível remover.")
            return None
        valor_removido = self.dados[self.inicio]
        self.inicio = (self.inicio + 1) % self.capacidade
        self.total -= 1
        return valor_removido

    def consultar_inicio(self):
        if self.esta_vazia():
            print("Fila vazia. Não há início.")
            return None
        return self.dados[self.inicio]

```

# Teste na main

```python
fila = FilaCircular(5)

# Inserções
fila.inserir(10)
fila.inserir(20)
fila.inserir(30)
fila.inserir(40)
fila.inserir(50)

# Tentativa de inserir com fila cheia
fila.inserir(60)

# Consulta
print("Início da fila:", fila.consultar_inicio())

# Remoção
print("Removido:", fila.remover())

# Inserção após remoção
fila.inserir(60)

# Impressão geral
while not fila.esta_vazia():
    print("Removido:", fila.remover())

print("Fila está vazia:", fila.esta_vazia())
print("Fila está cheia:", fila.esta_cheia())

```
